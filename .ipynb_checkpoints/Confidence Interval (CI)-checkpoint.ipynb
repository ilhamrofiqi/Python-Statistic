# Confidence Interval (CI)

Confidence Interval (CI) atau interval kepercayaan dalam statistik adalah rentang nilai yang digunakan untuk memperkirakan parameter populasi (seperti rata-rata atau proporsi), dan disertai dengan tingkat keyakinan (confidence level) tertentu.

 Komponen Confidence Interval:
- Titik estimasi (point estimate): nilai dari sampel, misalnya 𝑥 (rata-rata sampel)
- Margin of error (ME): seberapa jauh estimasi bisa meleset
- Tingkat kepercayaan (confidence level): biasanya 90%, 95%, atau 99%




```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Data sampel: tinggi badan siswa (dalam cm)
data = np.array([160, 165, 167, 170, 162, 164, 166, 168, 169, 171])

# Ukuran sampel, rata-rata, dan standar deviasi
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # ddof=1 agar std dev sampel

# Hitung margin of error (95% confidence)
confidence = 0.95
z_score = stats.norm.ppf((1 + confidence) / 2)  # Z untuk 95% = 1.96
margin_of_error = z_score * (std_dev / np.sqrt(n))

# Hitung confidence interval
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
# Visualisasi
plt.figure(figsize=(8, 2))
plt.errorbar(x=mean, y=0, xerr=margin_of_error, fmt='o', color='blue', capsize=5)
plt.axvline(mean, color='black', linestyle='--', label='Mean')
plt.title("95% Confidence Interval of Sample Mean")
plt.xlabel("Tinggi Badan (cm)")
plt.yticks([])
plt.legend()
plt.grid(True)
plt.show()
```
