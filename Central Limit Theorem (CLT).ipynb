# Central Limit Theorem (CLT)

Central Limit Theorem (CLT) atau Teorema Limit Tengah adalah salah satu konsep paling penting dalam statistik. CLT menjelaskan bahwa Jika kamu mengambil banyak sampel acak berukuran cukup besar dari sebuah populasi, maka distribusi rata-rata dari sampel-sampel itu akan mendekati distribusi normal, meskipun data populasi awalnya tidak normal.

Dalam Data Science, kita seringkali bekerja dengan dataset yang besar, model prediktif, dan kebutuhan untuk membuat inferensi dari data yang kita miliki. CLT adalah pilar teoretis yang mendukung banyak teknik dan asumsi yang kita gunakan.

Dan dalam dunia data science, kita jarang memiliki seluruh data populasi. Kita sering hanya punya sampel â€” entah itu log pengguna, data transaksi, atau polling. CLT memungkinkan data scientist untuk membuat inferensi (kesimpulan) dari sampel terhadap populasi secara andal.

```python
# Reload libraries and data since environment was reset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("employee.csv")
MonthlyIncome = df["MonthlyIncome"].dropna()

sample_means = []

# Ambil 1000 sampel, masing-masing terdiri dari 50 data acak
for _ in range(1000):
    sample = MonthlyIncome.sample(n=50, replace=True)
    sample_means.append(sample.mean())

# Plot histogram dari distribusi rata-rata sampel
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
plt.title("Central Limit Theorem Simulation\nDistribusi Rata-rata Sampel (n=50, 1000 kali)")
plt.xlabel("Sample Mean of MonthlyIncome")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Data populasi
populasi = [1000, 1200, 1500, 1600, 1700, 2000, 2500, 3000, 3500, 4000]

# Simulasi CLT: ambil 1000 sampel acak berukuran 3
sample_means = []
for _ in range(1000):
    sampel = np.random.choice(populasi, size=5, replace=True)
    sample_means.append(np.mean(sampel))

# Buat histogram dari rata-rata sampel
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title("Simulasi Central Limit Theorem (n=3, 1000 kali)")
plt.xlabel("Rata-rata Sampel")
plt.ylabel("Frekuensi")
plt.show()
```
