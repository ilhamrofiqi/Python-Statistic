# Measure of Standard Score

Dalam statistika, standard scores atau skor baku digunakan untuk menunjukkan seberapa jauh suatu nilai berada dari rata-rata (mean) dalam satuan standar deviasi. Ada beberapa jenis measures of standard scores, yang umum digunakan antara lain:

- Z-score: Menunjukkan berapa simpangan baku skor dari rata-rata. Rata-rata 0, simpangan baku 1. Paling dasar. <b>Rumus: Z = x-u/o</b>
- T-score: Versi Z-score yang dipermudah, tanpa negatif dan desimal. Rata-rata 50, simpangan baku 10. Dipakai di psikologi. <b>Rumus: T = 10 * z + 50

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("employee.csv")
df.head(3)
```

```python
# Calculate standard score (z-scores)
salary_zscores = stats.zscore(df['MonthlyIncome'])
df['MonthlyIncome_ZScore'] = salary_zscores
print('Standard Score (Z-Scores) of Salary:')
print(salary_zscores)

# Calculate standard score (t-scores)
salary_tscores = 10 * salary_zscores + 50
df['MonthlyIncome_TScore'] = salary_tscores
print('Standard Score (T-Scores) of Salary:')
print(salary_tscores)
```

```python
print("\nPlotting Z-Scores...")
df.plot(kind = 'line', x = 'MonthlyIncome', y = 'MonthlyIncome_ZScore')
plt.show()
```

```python
print("\nPlotting T-Scores...")
df.plot(kind = 'line', x = 'MonthlyIncome', y = 'MonthlyIncome_TScore', color='red')
plt.title('Monthly Income vs. T-Score')
plt.xlabel('Monthly Income')
plt.ylabel('T-Score')
plt.grid(True)
plt.show()
```
