# Measures of Shape and Position

1. Ukuran Bentuk (Measures of Shape)
Ukuran bentuk menggambarkan bagaimana data terdistribusi secara visual, terutama terkait simetri dan puncak.
- Kemencengan (Skewness)
Skewness mengukur tingkat ketidaksimetrisan distribusi data. Positif (menceng ke kanan). Negatif (menceng ke kiri). Nol (simetris)
- Keruncingan (Kurtosis)
Kurtosis mengukur "puncak" atau "keruncingan" distribusi data dibandingkan dengan distribusi normal (kurva lonceng). Ini juga melihat seberapa banyak data yang berada di "ekor" distribusi. Leptokurtik (Positif). Mesokurtik (Nol). Platykurtik (Negatif)

```python
import pandas as pd
import scipy.stats as stats
import numpy as np

df = pd.read_csv("employee.csv")
df.head(3)
```

```python
# Calculate skewness and kurtosis
salary_skewness = stats.skew(df["MonthlyIncome"])
salary_kurtosis = stats.kurtosis(df["MonthlyIncome"])

print("Skewness of Salary:", salary_skewness)
print("Kurtosis of Salary:", salary_kurtosis)
```

2. Ukuran Posisi (Measures of Position) Ukuran posisi menunjukkan lokasi relatif dari suatu nilai data dalam kumpulan data.
   -  Kuartil (Quartiles) Kuartil membagi data yang telah diurutkan menjadi empat bagian yang sama. Q1 = 25%. Q2 = Median. Q3 = 75%
   -  Persentil (Percentiles) Persentil membagi data yang telah diurutkan menjadi 100 bagian yang sama.

```python
# Calculate percentiles
percentiles = np.percentile(df['MonthlyIncome'], [15, 45, 65])

#Calculate quartiles
quartiles = np.percentile(df['MonthlyIncome'], [25, 50, 75])

print('Percentiles:')
print('15 percentile:', percentiles[0])
print('45 percentile (median):', percentiles[1])
print('65 percentile:', percentiles[2])

print('\nQuartiles:')
print('Q1 (25th percentile):', quartiles[0])
print('Q1 (50th percentile, median):', quartiles[1])
print('Q1 (75th percentile):', quartiles[2])
```
