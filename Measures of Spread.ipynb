# Measures of Spread

Ukuran penyebaran data (Measures of Spread, juga dikenal sebagai measures of dispersion atau variability) adalah nilai statistik yang menggambarkan seberapa tersebar atau bervariasinya data dalam suatu kumpulan. Ini melengkapi ukuran pemusatan data (seperti mean, median, modus) yang hanya menunjukkan "pusat" data.

Berikut adalah beberapa ukuran penyebaran data yang umum:
1. Rentang (Range)
Rentang adalah ukuran penyebaran yang paling sederhana, dihitung sebagai selisih antara nilai terbesar (maksimum) dan nilai terkecil (minimum) dalam kumpulan data.
2. Variansi (Variance)
Variansi mengukur seberapa jauh setiap titik data dari mean (rata-rata) dan seberapa jauh titik-titik data itu saling berjauhan. Ini dihitung dengan merata-ratakan kuadrat dari selisih setiap titik data dengan mean.
3. Standar Deviasi (Standard Deviation - SD)
Standar deviasi adalah akar kuadrat dari variansi. Ini adalah ukuran penyebaran yang paling umum digunakan karena memiliki satuan yang sama dengan data asli, sehingga lebih mudah diinterpretasikan. Standar deviasi menunjukkan seberapa jauh, rata-rata, setiap titik data menyimpang dari mean.

```python
import pandas as pd

df = pd.read_csv("employee.csv")
df.head()
```

```python
salary_range = df["MonthlyIncome"].max() - df["MonthlyIncome"].min()
salary_variance = df["MonthlyIncome"].var()
salary_std_dev = df["MonthlyIncome"].std()

print('Measures of Spread - MonthlyIncome:')
print('Salary Range:', salary_range)
print('Salary Variance:', salary_variance)
print('Salary Standard Deviation:', salary_std_dev)
```
