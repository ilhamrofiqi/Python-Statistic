# Normal Distribution

Distribusi normal, juga dikenal sebagai distribusi Gaussian atau kurva lonceng. Ini menggambarkan distribusi probabilitas kontinu yang simetris di sekitar rata-ratanya, di mana data di dekat rata-rata lebih sering muncul daripada data yang jauh dari rata-rata.

### Berikut adalah karakteristik dan properti utama dari distribusi normal:
##### - Aturan Empiris (Aturan 68-95-99.7): Aturan ini menggambarkan persentase data yang jatuh dalam deviasi standar tertentu dari rata-rata:
- Sekitar 68% data jatuh dalam satu deviasi standar (±1σ) dari rata-rata.
- Sekitar 95% data jatuh dalam dua deviasi standar (±2σ) dari rata-rata.
- Sekitar 99.7% data jatuh dalam tiga deviasi standar (±3σ) dari rata-rata.

##### - Kurva Berbentuk Lonceng
##### - Simetri
##### Dan lain lain (lengkap cari di Gemini)

### Mengapa Ini Penting?
Distribusi normal digunakan secara luas di banyak bidang karena:

- Fenomena Alam: Banyak fenomena alam dan sosial cenderung mengikuti distribusi normal atau mendekati normal (misalnya, tinggi badan manusia, tekanan darah, nilai ujian, kesalahan pengukuran).
- Teorema Batas Pusat (CLT): Teorema fundamental ini menyatakan bahwa distribusi rata-rata sampel dari sejumlah besar variabel acak yang independen dan terdistribusi secara identik akan kira-kira normal, terlepas dari distribusi asli variabel tersebut. Ini menjadikan distribusi normal sangat penting untuk statistika inferensial.
- Inferensi Statistik: Ini berfungsi sebagai dasar untuk berbagai uji statistik dan interval kepercayaan, memungkinkan kita membuat kesimpulan tentang populasi berdasarkan data sampel.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("employee.csv")
df.head()
```

```python
MonthlyIncome = df["MonthlyIncome"]

plt.hist(MonthlyIncome, bins=30)
plt.xlabel('MonthlyIncome')
plt.ylabel('Frequency')
plt.title('Distribution of Salary')
plt.show()

mean = MonthlyIncome.mean()
std_dev = MonthlyIncome.std()
print("Mean:{0:.2f}". format(mean))
print("Standart Deviation:{0:.2f}". format(std_dev))
```

```python
from scipy.stats import norm

# Fit a normal distribution
mu, sigma = norm.fit(MonthlyIncome)

# Ploting the histogram and fitted normal distribution
plt.hist(MonthlyIncome, bins=30, density=True, alpha=0.6)
x = np.linspace(min(MonthlyIncome), max(MonthlyIncome), 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'r-', linewidth=2)

plt.xlabel('MonthlyIncome')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Salary')
plt.show()

print('Mean:{0:.2f}'.format(mean))
print('Standart Deviation:{0:.2f}'.format(std_dev))
```
