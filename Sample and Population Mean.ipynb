# Sample and Polulation Mean

### Population
Populasi adalah keseluruhan kelompok atau koleksi semua individu, objek, peristiwa, atau pengamatan yang ingin Anda tarik kesimpulan atau pelajari. Ini adalah kelompok target yang lengkap.

Contoh:
- Semua mahasiswa di sebuah universitas.
- Semua pohon pinus di hutan tertentu.
- Semua produk yang dihasilkan oleh sebuah pabrik dalam satu hari.

### Population Mean
Rata-rata populasi adalah rata-rata aritmatika dari semua nilai dalam keseluruhan populasi. Ini adalah nilai "benar" atau parameter sebenarnya dari populasi yang seringkali ingin kita ketahui. Dilambangkan dengan huruf Yunani μ (mu).



### Sample
Sampel adalah subset atau bagian yang lebih kecil dari populasi yang dipilih untuk studi. Data dikumpulkan dari sampel ini.

Contoh:
- 100 mahasiswa yang dipilih secara acak dari sebuah universitas.
- 50 pohon pinus yang dipilih dari hutan.
- 10 produk yang dipilih dari lini produksi pabrik.

### Sample Mean
 Rata-rata sampel adalah rata-rata aritmatika dari semua nilai dalam sampel. Ini adalah estimasi dari rata-rata populasi. Dilambangkan dengan xˉ(x-bar).

 

```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv("employee.csv")
df.head()
```

```python
# The population data is the complete data
population_data = sorted(df['MonthlyIncome'])

# The sample data is a sample of the complete data
# let's set the sample size to be 150
sample_data = random.sample(population_data, 150)

# Population mean
print('Population Mean:', np.mean(population_data))

# Sample Mean
print('Sample Mean:', np.mean(sample_data))
```
