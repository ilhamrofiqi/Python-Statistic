# Statistical Transformations

Transformasi statistik, juga dikenal sebagai transformasi data, adalah operasi matematika yang diterapkan pada data untuk mengubah distribusinya, mengurangi kemencengan, atau mencapai properti lain yang diinginkan. Transformasi ini umumnya digunakan dalam ilmu data karena berbagai alasan. Mari kita jelajahi beberapa jenis spesifik transformasi statistik dan mengapa itu penting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
df = pd.read_csv("employee.csv")
df.head()
```

```python
# Original data visualization
MonthlyIncome = df["MonthlyIncome"]

plt.hist(MonthlyIncome, bins=30, edgecolor='black')
plt.title("MonthlyIncome Data")
plt.xlabel("MonthlyIncome")
plt.ylabel("Frequency")
plt.show()
```

 ### 4 transformasi statistik yang paling umum digunakan:


#### 1. Square Root Transformation (Transformasi Akar Kuadrat)

Transformasi akar kuadrat melibatkan pengambilan akar kuadrat dari setiap titik data. Ini berguna untuk mengurangi kemencengan kanan (right skewness) dari distribusi yang miring positif. Dengan mengompres nilai-nilai yang lebih besar lebih dari nilai-nilai yang lebih kecil, ini membantu menormalkan data dan membuatnya lebih simetris.

Fungsi: Mengurangi skewness (kemencengan) pada data dengan nilai kecil dan varian besar.

Rumus:


 
Contoh Kasus: Data hitungan (count), seperti jumlah pelanggan, jumlah cacat produksi, dll.

```python
# Square Root Transformation

plt.hist(np.sqrt(MonthlyIncome), bins=30, edgecolor='black')

plt.title('Square Root Transformation')
plt.xlabel('Square Root MonthlyIncome')
plt.ylabel('Frequency')
plt.show()
print("Sebelum transformasi:", skew(MonthlyIncome))
print("Setelah sqrt transformasi:", skew(np.sqrt(MonthlyIncome)))
```

#### 2. Cube Root Transformation (Transformasi Akar Pangkat Tiga)

Transformasi akar kubik melibatkan pengambilan akar kubik dari setiap titik data. Ini juga membantu dalam mengurangi kemencengan positif, tetapi pada tingkat yang lebih besar dibandingkan dengan transformasi akar kuadrat. Transformasi ini bisa sangat efektif ketika berhadapan dengan data yang mengandung nilai positif dan negatif.

Fungsi: Digunakan untuk data dengan nilai negatif dan positif karena akar pangkat tiga bisa diterapkan pada bilangan negatif.

Rumus:


 
Contoh Kasus: Data keuangan, pertumbuhan, atau perbedaan nilai yang bisa positif atau negatif.

```python
# Cube Root Transformation

plt.hist(np.power(MonthlyIncome, 1/3), bins=30, edgecolor='black')
plt.title('Cube Root Transformation')
plt.xlabel('Cube Root Transformed MonthlyIncome')
plt.ylabel('Frequency')
plt.show()
print("Sebelum transformasi:", skew(MonthlyIncome))
print("Setelah Cube Root transformasi:", skew(np.power(MonthlyIncome, 1/3)))
```

#### 3. Log Transformation (Transformasi Logaritmik)

Transformasi logaritma melibatkan pengambilan logaritma dari setiap titik data. Ini umumnya digunakan ketika data memiliki distribusi yang sangat miring (highly skewed), terutama dengan ekor kanan yang panjang. Transformasi logaritma mengompres nilai-nilai besar lebih dari nilai-nilai yang lebih kecil, mengurangi rentang dan membuat distribusi lebih simetris.

Fungsi: Mengurangi skewness kanan (positif) dan menstabilkan varians.

Rumus:



Catatan: Hanya bisa digunakan pada data positif (> 0).

Contoh Kasus: Data penjualan, pendapatan, harga rumah, dll.

```python
# Log Transformation

plt.hist(np.log(MonthlyIncome), bins=30, edgecolor='black')
plt.title("Log Transformation")
plt.xlabel('Log MonthlyIncome')
plt.ylabel('Frequency')
plt.show()
print("Sebelum transformasi:", skew(MonthlyIncome))
print("Setelah Log transformasi:", skew(np.log(MonthlyIncome)))
```

#### 4. Box-Cox Transformation
Transformasi Box-Cox adalah teknik transformasi yang lebih umum yang mencakup transformasi akar kuadrat dan logaritma sebagai kasus-kasus khusus. Ini dirancang untuk menemukan parameter transformasi pangkat terbaik, lambda (Î»), untuk mencapai properti distribusi yang paling diinginkan. Transformasi Box-Cox dapat menangani berbagai jenis kemencengan dan dapat diterapkan pada nilai positif dan negatif.

Fungsi: Transformasi fleksibel untuk membuat data lebih normal (distribusi mendekati normal).

Rumus umum:


 
Kelebihan: Mencari nilai optimal Î» (lambda) untuk transformasi terbaik.

Catatan: Hanya untuk data positif.

```python
from scipy import stats

# Box-Cox transformation
boxcox_MonthlyIncome, _= stats.boxcox(MonthlyIncome)
plt.hist(boxcox_MonthlyIncome, bins=30, edgecolor='black')
plt.title('Box-Cox Transformation')
plt.xlabel('Box-Cox Transformed MonthlyIncome')
plt.ylabel('Frequency')
plt.show()
print("Sebelum transformasi:", skew(MonthlyIncome))
print("Setelah Box-Cox transformasi:", skew(boxcox_MonthlyIncome))
```
